## **`Multivariate Plots`**

Multivariate analysis (MVA) is a Statistical procedure for analysis of data involving more than one type of measurement or observation. It may also mean solving problems where more than one dependent variable is analyzed simultaneously with other variables.

## **`1. Stacked Bar Plot`**

A stacked chart is a form of bar chart that shows the composition and comparison of a few variables, either relative or absolute, over time. Also called a stacked bar or column chart, they look like a series of columns or bars that are stacked on top of each other.

```python
import matplotlib.pyplot as plt
from palettable.colorbrewer.qualitative import Pastel1_7

# Define data
labels = ['Group A', 'Group B', 'Group C', 'Group D']
sizes = [12, 11, 3, 30]

# Create a pie chart
colors = Pastel1_7.hex_colors  # Use pastel colors from the palette
plt.pie(sizes, labels=labels, colors=colors)

# Add a white circle to create a donut chart
center_circle = plt.Circle((0, 0), 0.7, color='white')
plt.gca().add_artist(center_circle)

# Display the chart
plt.axis('equal')  # Ensure the pie chart is a circle
plt.show()
```

```python
# # libraries
import matplotlib.pyplot as plt
import squarify    # pip install squarify (algorithm for treemap)
import pandas as pd

# Create a data frame with sample data
df = pd.DataFrame({'nb_people':[8,3,4,2], 'group':["group A", "group B", "group C", "group D"] })

# plot it
squarify.plot(sizes=df['nb_people'], label=df['group'], alpha=.8 )
plt.axis('off')
plt.show()
```

```python
# Import the pandas library
import pandas as pd

# Make a data frame with dots to show on the map
data = pd.DataFrame({
   'lon':[-58, 2, 145, 30.32, -4.03, -73.57, 36.82, -38.5],
   'lat':[-34, 49, -38, 59.93, 5.33, 45.52, -1.29, -12.97],
   'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador'],
   'value':[10, 12, 40, 70, 23, 43, 100, 43]
}, dtype=str)

data
```

```python
# import the library
import folium

# Make an empty map
m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)

# Show the map
m
```

```python
for i in range(0,len(data)):
   folium.Marker(
      location=[data.iloc[i]['lat'], data.iloc[i]['lon']],
      popup=data.iloc[i]['name'],
   ).add_to(m)

# Show the map again
m
```
